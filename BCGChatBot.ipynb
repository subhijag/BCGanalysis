```python
import pandas as pd

df = pd.read_csv('/content/Company_Financials_2021_2023.csv')
df
```

```python
def simple_chatbot(user_query):
    if user_query == "What is the total revenue of Apple in 2023?":
        return "Apple's total revenue in 2023 was $383,285 million."

    elif user_query == "How has Tesla's net income changed from 2022 to 2023?":
        change = 14997 - 12556
        return f"Tesla's net income increased by ${change} million from 2022 to 2023."

    elif user_query == "What are Microsoft’s total assets in 2021?":
        return "Microsoft’s total assets in 2021 were $333,779 million."

    elif user_query == "How much was the operating cash flow of Apple in 2022?":
        return "Apple's operating cash flow in 2022 was $122,151 million."

    elif user_query == "What is Tesla’s total liability in 2023?":
        return "Tesla’s total liabilities in 2023 were $40,553 million."

    else:
        return "Sorry, I can only provide information on predefined financial queries."


# Simple command-line loop to test the chatbot
if __name__ == "__main__":
    print("Welcome to the Financial Chatbot. Type 'exit' to quit.")
    while True:
        query = input("\nAsk a question: ")
        if query.lower() == "exit":
            break
        response = simple_chatbot(query)
        print("Chatbot:", response)
```
