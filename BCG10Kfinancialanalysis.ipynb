```python
import pandas as pd

df = pd.read_csv('/content/Company_Financials_2021_2023.csv')
df.head()
```

```python
# Calculate YoY percentage growth for revenue and net income
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue (USD Millions)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income (USD Millions)'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets (USD Millions)'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities (USD Millions)'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Operating Cash Flow (USD Millions)'].pct_change() * 100

# Display the updated DataFrame
df
```

###  Summary of Financial Trends (20212023)

- **Apple** showed steady growth in revenue until 2022, followed by a slight decline in 2023. Net income remained strong but also declined slightly in the latest year.
- **Tesla** experienced strong revenue and net income growth across all years, indicating rapid expansion and improving profitability.
- **Microsoft** maintained stable revenue growth and strong net income. Operating cash flow steadily increased, indicating healthy cash generation.

Overall, all three companies show strong financials, with Tesla growing fastest and Apple maintaining the highest income and cash flows.
